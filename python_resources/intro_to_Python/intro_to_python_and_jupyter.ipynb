## Jupyter Notebook Basics

### Two Modes

- command mode
    - press escape to enter command mode
- editting
    - click on a cell or press enter to start editting mode
 
### Two Main Cell Types

- Python code cells
- Markdown cells for formatted text
    - Markdown is a simple markup language that allows headings, bullet points, equations, and so on

### Executing Python Code

- you can put multiple lines of Python code in one cell
- pressing enter while typing code will go to the next line
- press shift+enter to execute the code
- if the last line of code returns a result that is not captured by a variable, it will print out below the exectute cell
