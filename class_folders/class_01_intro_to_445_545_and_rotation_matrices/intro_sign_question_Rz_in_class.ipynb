## Issues

- sign on $Y$ term is wrong
- `robotics.Rz` gives $^A_B\mathbf{R}$
- we need $^B_A\mathbf{R}$
    - because we know $^AP$
 
$$^BP = \; ^B_A\mathbf{R} \cdot ^AP$$
